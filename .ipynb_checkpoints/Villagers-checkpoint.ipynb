```python
# Dependencies for analysis
import numpy as np 
import pandas as pd 

# Dependencies for graphing
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
# Read in .csv file
villager_info=pd.read_csv('Resources/villagers.csv')
```

```python
# Make a dataframe
villagers_df=pd.DataFrame(villager_info)

villagers_df
```

```python
# Find NaN and drop it
villagers_df.dropna()
```

```python
# Drop columns that are not needed
villagers_df.drop(columns=['Name', 'Gender', 'Hobby', 'Catchphrase', 'Style 1', 'Style 2', 'Color 1', 'Color 2',
                          'Wallpaper', 'Flooring', 'Furniture List', 'Filename', 'Unique Entry ID', 'Birthday'], inplace=True)

villagers_df.head()
```

```python
# How many of each species
species_list=villagers_df.value_counts('Species')

species_count_df=pd.DataFrame(species_list)

species_count_df.head()
```

```python
# Rename 0 column
species_count_df.rename(columns={0:'Count of Species'}, inplace=True)

species_count_df.head()
```

```python
# Plot count of each species
plt.figure(figsize=(50,50))

sns.color_palette('rocket', as_cmap=True)

sns.barplot(x=species_count_df.index, y='Count of Species', data=species_count_df,
           palette='rocket')

plt.title('Species of Villagers in Animal Crossing', fontsize=75)

plt.xlabel('Species', fontsize=70)

plt.ylabel('Count of Species', fontsize=70)

plt.xticks(rotation=70, fontsize=55)

plt.yticks(fontsize=50)
```

```python
# Type of personalities
villagers_df.head()

personality_list=villagers_df.value_counts('Personality')

personality_list
```

```python
# Create a dataframe
personality_df=pd.DataFrame(personality_list)

personality_df.head()
```

```python
# Rename columns
personality_df.rename(columns={0:'Count of Personalities'}, inplace=True)

personality_df.head()
```

```python
# Barchart for personalities
plt.figure(figsize=(20,15),dpi=80)

sns.barplot(x=personality_df.index, y='Count of Personalities', data=personality_df,
           palette='magma')

plt.title('Personalities of Villagers in Animal Crossing', fontsize=50)

plt.xlabel('Personalities', fontsize=40)

plt.ylabel('Count of Personalities', fontsize=40)

plt.xticks(rotation=60, fontsize=25)

plt.yticks(fontsize=25)
```

```python
# Favorite songs for villagers
favorite_songs=villagers_df.value_counts('Favorite Song')

favorite_songs
```

```python
# Make dataframe for villager's favorite songs
favorite_songs_df=pd.DataFrame(favorite_songs)

favorite_songs_df.head()
```

```python
# Rename column
favorite_songs_df.rename(columns={0:'Count of Favorite Songs'}, inplace=True)

favorite_songs_df.head()
```

```python
# Bar chart
plt.figure(figsize=(35,25), dpi=80)

sns.barplot(x=favorite_songs_df.index, y='Count of Favorite Songs', data=favorite_songs_df,
           palette='magma')

plt.title('Favorite Songs of Villagers in Animal Crossing', fontsize=80)

plt.xlabel('Song Title', fontsize=50)

plt.ylabel('Count of Favorite Song', fontsize=50)

plt.xticks(rotation=90, fontsize=25)

plt.yticks(fontsize=35)
```
